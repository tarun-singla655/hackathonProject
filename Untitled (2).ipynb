```python
dummies1 =  pd.get_dummies(test_data.car_ownership)
dummies =  pd.get_dummies(train_data.car_ownership)
train_data = pd.concat([train_data,dummies],axis = 'columns')
test_data = pd.concat([test_data,dummies1],axis = 'columns')
train_data = train_data.drop(['car_ownership'],axis = 1)
test_data = test_data.drop(['car_ownership'],axis = 1)
dummies1 =  pd.get_dummies(test_data.city)
dummies =  pd.get_dummies(train_data.city)
train_data = pd.concat([train_data,dummies],axis = 'columns')
test_data = pd.concat([test_data,dummies1],axis = 'columns')
train_data = train_data.drop(['city'],axis = 1)
test_data = test_data.drop(['city'],axis = 1)
dummies1 =  pd.get_dummies(test_data.profession)
dummies =  pd.get_dummies(train_data.profession)
train_data = pd.concat([train_data,dummies],axis = 'columns')
test_data = pd.concat([test_data,dummies1],axis = 'columns')
train_data = train_data.drop(['profession'],axis = 1)
test_data = test_data.drop(['profession'],axis = 1)
```

```python
dummies1 =  pd.get_dummies(test_data.state)
dummies =  pd.get_dummies(train_data.state)
```

```python
train_data = pd.concat([train_data,dummies],axis = 'columns')
test_data = pd.concat([test_data,dummies1],axis = 'columns')
```

```python
train_data = train_data.drop(['Id'],axis = 1)
test_id = test_data['id']
test_data = test_data.drop(['id'] , axis = 1)
```

```python
Y_train = train_data['risk_flag']
```

```python
X_train = train_data.drop(['risk_flag'] ,axis = 1)
```

```python
X_test = test_data
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential([
    layers.Dense(407,activation = 'relu'),
    layers.Dense(302,activation = 'relu'),
    layers.Dense(261,activation = 'relu'),
    layers.Dense(200,activation = 'relu'),
    layers.Dense(152,activation = 'relu'),
    layers.Dense(101,activation = 'relu'),
      layers.Dense(71,activation = 'relu'),
      layers.Dense(23,activation = 'relu'),
      layers.Dense(10,activation = 'relu'),
      layers.Dense(4,activation = 'relu'),
    layers.Dense(1,activation = 'sigmoid'),
])
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'],
)
model.fit(X_train,Y_train,epochs = 100)
```

```python
y = model.predict(X_test)
```

```python
result = []
for i in y:
    if i < 0.5:
        result.append(0);
    else:
        result.append(1);
```

```python
import numpy as np
res = np.array(result)
```
